```python
average=ana[['Year',"Month",'Y','actual-y']]
average.head()
```

```python
average['log']=np.log(average['actual-y'])
```

```python
average.head()
```

```python
average['date'] = pd.to_datetime(average[['Year', 'Month']].astype(str).agg('-'.join, axis=1))
```

```python
average['date'] = pd.PeriodIndex(year=average['Year'], month=average['Month'], freq='M')
```

```python
average.head()
```

```python
# Assuming 'average' is a DataFrame and you want to perform the test on the 'log' column
result = adfuller(average['actual-y'], autolag='AIC')

print(f'ADF Statistic: {result[0]}')
print(f'n_lags: {result[1]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critical Values:')
    print(f'   {key}, {value}')
```

```python
average.set_index('date', inplace=True)
```

```python
average.head()
```

```python
result_add = seasonal_decompose(average['actual-y'], model='additive',period=4)
```

```python
fig, axes = plt.subplots(4, 1, figsize=(10, 8), sharex=True)
result_add.observed.plot(ax=axes[0], legend=False)
axes[0].set_ylabel('Observed')
result_add.trend.plot(ax=axes[1], legend=False)
axes[1].set_ylabel('Trend')
result_add.seasonal.plot(ax=axes[2], legend=False)
axes[2].set_ylabel('Seasonal')
result_add.resid.plot(ax=axes[3], legend=False)
axes[3].set_ylabel('Residual')
plt.show()
```

# ACF & PACF

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from pandas.plotting import autocorrelation_plot
```

```python
autocorrelation_plot(average[['actual-y']])
plt.show()
```

```python
import statsmodels.api as sm
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(average['actual-y'].dropna(),lags=15,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(average['actual-y'].dropna(),lags=15,ax=ax2)
```

# ARIMA

```python
from statsmodels.tsa.arima.model import ARIMA
model=ARIMA(average['actual-y'],order=(1,1,1))
model_fit=model.fit()
model_fit.summary()
```

```python
average['forecast']=model_fit.predict(start=1,end=42)
average[['actual-y','forecast']].plot(figsize=(12,8))
```

```python
import statsmodels.api as sm
model=sm.tsa.statespace.SARIMAX(average['actual-y'],order=(1, 1, 1),seasonal_order=(1,1,1,4))
results=model.fit()
average['forecast']=results.predict(start=30,end=37,dynamic=True)
average[['actual-y','forecast']].plot(figsize=(12,8))
```

```python
from pandas.tseries.offsets import DateOffset
future_dates=[average.index[-1]+ DateOffset(months=x)for x in range(0,5)]
future_datest_df=pd.DataFrame(index=future_dates[1:],columns=df.columns)

future_datest_df.tail()

future_df=pd.concat([average,future_datest_df])

future_df['forecast'] = results.predict(start =37 , end = 41, dynamic= True)
future_df[['actual-y', 'forecast']].plot(figsize=(12, 8))
```
