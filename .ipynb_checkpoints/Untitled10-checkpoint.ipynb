```python
flagyl['log_y'] = np.log(flagyl['actual_y'])
```

```python
flagyl['actual_y'] = flagyl['y'] - flagyl['received']
```

```python
sns.lineplot(data=flagyl, x='month', y='actual_y', hue='year', marker='o')
```

```python

pivot_table = flagyl.pivot(index='year', columns='month', values='actual_y')

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(10, 10))


theta = np.linspace(0.0, 2 * np.pi, num=12, endpoint=False)
theta = np.concatenate((theta, [theta[0]])) 

for year in pivot_table.index:
    radii = pivot_table.loc[year].values
    radii = np.concatenate((radii, [radii[0]])) 
    ax.plot(theta, radii, label=str(year))


ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_xticks(np.linspace(0, 2 * np.pi, 12, endpoint=False))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

ax.set_ylabel('$ million')
ax.set_title('Flagyl Tab Sales')

ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

plt.show()
```

```python

result_add = seasonal_decompose(ts['log_y'], model='additive')
```

```python
plt.figure(figsize=(10, 4))
plt.plot(result_add.seasonal, label='Additive Seasonal')
plt.legend()
```

```python

plt.figure(figsize=(10, 8))
plt.plot(result_add.trend, label='Additive trend')
plt.legend()
```

```python
fig, axes = plt.subplots(4, 1, figsize=(10, 8), sharex=True)
result_add.observed.plot(ax=axes[0], legend=False)
axes[0].set_ylabel('Observed')
result_add.trend.plot(ax=axes[1], legend=False)
axes[1].set_ylabel('Trend')
result_add.seasonal.plot(ax=axes[2], legend=False)
axes[2].set_ylabel('Seasonal')
result_add.resid.plot(ax=axes[3], legend=False)
axes[3].set_ylabel('Residual')
plt.show()
```

```python
   
plt.figure(figsize=(9, 3))
plt.plot(ts.log_y, label='Original Time Series')
plt.plot(result_add.trend, label='-trend')
plt.legend()
```

```python
decomposition = sm.tsa.x13_arima_analysis(ts['actual_y'])
```

```python
trend = decomposition.trend
trend = decomposition.seasadj  # Seasonally adjusted series (seasonal component)
seasonal = decomposition.observed  # Observed series (original series)
irregular = decomposition.irregular
```

```python
fig, axes = plt.subplots(4, 1, figsize=(12, 10), sharex=True)
seasonal.plot(ax=axes[0], legend=False)
axes[0].set_ylabel('Observed')
trend.plot(ax=axes[1], legend=False)
axes[1].set_ylabel('Trend')
decomposition.seasonal.plot(ax=axes[2], legend=False)
axes[2].set_ylabel('Seasonal')
irregular.plot(ax=axes[3], legend=False)
axes[3].set_ylabel('Irregular')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.seasonal import STL
```

```python
stl = STL(ts['actual_y'], seasonal=13)
```

```python
result = stl.fit()
```

```python
trend = result.trend
seasonal = result.seasonal
residual = result.resid
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(ts['actual_y'], label='Original', color='blue')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend, label='Trend', color='red')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal, label='Seasonal', color='green')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual, label='Residual', color='purple')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
