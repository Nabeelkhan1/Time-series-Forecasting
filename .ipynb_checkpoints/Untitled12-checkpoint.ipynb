```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('/home/hamza/Downloads/saadHospitalPharmacy.csv')
```

```python
df.columns = ['hospital', 'store', 'year', 'month','item', 'transaction', 'y', 'received']
```

```python
df
```

```python
panadol = df[df['item'] == 'Tab Panadol  (Paracetamol) 500mg']
```

```python
ts.set_index('date', inplace=True)
```

# Time Series Data Visualization
We took panadol time series from orignal data we are going to visualize data below:

```python
sns.lineplot(data=panadol, x='month', y='actual_y', hue='year', marker='o')
```

```python

pivot_table = panadol.pivot(index='year', columns='month', values='actual_y')

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(10, 10))


theta = np.linspace(0.0, 2 * np.pi, num=12, endpoint=False)
theta = np.concatenate((theta, [theta[0]])) 

for year in pivot_table.index:
    radii = pivot_table.loc[year].values
    radii = np.concatenate((radii, [radii[0]])) 
    ax.plot(theta, radii, label=str(year))


ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_xticks(np.linspace(0, 2 * np.pi, 12, endpoint=False))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

ax.set_ylabel('$ million')
ax.set_title('Panadol Tab Sales')

ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
    
plt.show()
```

# Time Series Stationary Or Non Stationary
If p-value is less then 0.05 -> time series stationary
If p-value is greater then 0.05 -> time series non-stationary
Dicky-Fuller Test to check whether series is stationary or not

```python
from statsmodels.tsa.stattools import adfuller, kpss
```

```python
adf_result = adfuller(ts)
```

```python
print('p-value:', adf_result[1])
```

# Time Series Decomposition
In time series decomposition we separate Trend, Seasonality and residual. <br>
Additive Decomposition if data has constant mean, variance and Autocorr. <br>
* yt = St + Tt + Rt * <br>
Multiplicative Decomposition if mean, variance and autocorr varie. <br>
* yt = St * Tt * Rt * <br>
Models For Decomposition. <br>
- item Classical Decompsition
- item X11 Decompsition
- item SEAT Decompsition
- item STL Decompsition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm
```

```python
result_add = seasonal_decompose(ts['actual_y'], model='additive', period = 12)
```

```python
fig, axes = plt.subplots(4, 1, figsize=(10, 8), sharex=True)
result_add.observed.plot(ax=axes[0], legend=False)
axes[0].set_ylabel('Observed')
result_add.trend.plot(ax=axes[1], legend=False)
axes[1].set_ylabel('Trend')
result_add.seasonal.plot(ax=axes[2], legend=False)
axes[2].set_ylabel('Seasonal')
result_add.resid.plot(ax=axes[3], legend=False)
axes[3].set_ylabel('Residual')
plt.show()
```

# Issues in Classical decomposition above

- item Seasonality constant at every interval or year

# Exponential Smoothing
Now we going to apply simple exponential smoothing on time series data.<br>
Let's start with simple exponential smoothing.
### Simple Exponential Smoothing (SES)

```python
from statsmodels.tsa.api import SimpleExpSmoothing
import plotly.express as px
```

```python
smoothing_level = 0.8
```

```python
ses = SimpleExpSmoothing(ts)
```

```python
ts
```

```python
split_date = '2024-01-01'
split_date = pd.to_datetime(split_date)
```

```python
train_data = ts[ts.index <= split_date]
```

```python
predict_data = ts[ts.index > split_date]
```

```python
ts.plot(marker='o', color='black', legend=True, figsize=(14, 7))
fit1 = SimpleExpSmoothing(train_data).fit(smoothing_level=smoothing_level,optimized=False)
fcast1 = fit1.forecast(len(predict_data)).rename(r'$\alpha={}$'.format(smoothing_level))
fcast1.plot(marker='o', color='blue', legend=True)
fit1.fittedvalues.plot(marker='o',  color='blue')
mse1 = ((fcast1 - predict_data.actual_y) ** 2).mean()
print('The Root Mean Squared Error of our forecasts with auto optimization is {}'.format(round(np.sqrt(mse1), 2)))
```

```python
ts.plot(marker='o', color='black', legend=True, figsize=(14, 7))
fit2 = SimpleExpSmoothing(train_data).fit()
fcast2 = fit2.forecast(len(predict_data)).rename(r'$\alpha=%s$'%fit2.model.params['smoothing_level'])
fcast2.plot(marker='o', color='green', legend=True)
fit2.fittedvalues.plot(marker='o', color='green')
mse2 = ((fcast2 - predict_data.actual_y) ** 2).mean()
print('The Root Mean Squared Error of our forecasts with auto optimization is {}'.format(round(np.sqrt(mse2), 2)))
plt.show()
```

```python
ts.plot(marker='o', color='black', legend=True, figsize=(14, 7))
fcast1.plot(marker='o', color='blue', legend=True)
fit1.fittedvalues.plot(marker='o',  color='blue')
print('The Root Mean Squared Error of our forecasts with out auto optimization is {}'.format(round(np.sqrt(mse1), 2)))
fcast2.plot(marker='o', color='green', legend=True)
fit2.fittedvalues.plot(marker='o', color='green')
print('The Root Mean Squared Error of our forecasts with auto optimization is {}'.format(round(np.sqrt(mse2), 2)))
```

```python
ts.plot(marker='o', color='black', legend=True, figsize=(14, 7))
fit2 = Holt(train_data, exponential=True).fit(smoothing_level, smoothing_trend, optimized=False)
fcast2 = fit2.forecast(len(predict_data)).rename("Exponential trend")
mse2 = ((fcast2 - predict_data.actual_y) ** 2).mean()
print('The Root Mean Squared Error of Holt''s Exponential trend {}'.format(round(np.sqrt(mse2), 2)))

fit2.fittedvalues.plot(marker="o", color='red')
fcast2.plot(color='red', marker="o", legend=True)

plt.show()
```

# Holt Winter Seasonal Additive Method
In previous models we only tackle with trend in data. Now in Holt winter method we are handle seaonal component in 
series.<br>



```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
seasonal_period = 12
ts.plot(marker='o', color='black', legend=True, figsize=(14, 7))
fit1 = ExponentialSmoothing(train_data, seasonal_periods = seasonal_period, trend='add', seasonal='add', use_boxcox=True).fit()
fcast1 = fit1.forecast(len(predict_data)).rename('Additive')
mse1 = ((fcast1 - predict_data.actual_y) ** 2).mean()
print('The Root Mean Squared Error of additive trend, additive seasonal of '+ 
      'period season_length={} and a Box-Cox transformation {}'.format(seasonal_period,round(np.sqrt(mse1), 2)))

fit2 = ExponentialSmoothing(train_data, seasonal_periods = seasonal_period, trend='add', seasonal='add', damped=True, use_boxcox=True).fit()
fcast2 = fit2.forecast(len(predict_data)).rename('Additive+damped')
mse2 = ((fcast2 - predict_data.actual_y) ** 2).mean()
print('The Root Mean Squared Error of additive damped trend, additive seasonal of '+ 
      'period season_length={} and a Box-Cox transformation {}'.format(seasonal_period,round(np.sqrt(mse2), 2)))

fit1.fittedvalues.plot(style='--', marker = 'o', color='red')
fcast1.plot(style='--', marker='o', color='red', legend=True)
fit2.fittedvalues.plot(style='--', marker = 'o', color='green')
fcast2.plot(style='--', marker='o', color='green', legend=True)
```

```python
ts.plot(marker='o', color='black', legend=True, figsize=(14, 7))
fit3 = ExponentialSmoothing(train_data, seasonal_periods = seasonal_period, trend='add', seasonal='mul',use_boxcox=True).fit()
fcast3 = fit3.forecast(len(predict_data)).rename('Multiplicative')
mse3 = ((fcast3 - predict_data.actual_y) ** 2).mean()
print('The Root Mean Squared Error of additive trend, multiplicative seasonal of '+ 
      'period season_length={} and a Box-Cox transformation {}'.format(seasonal_period,round(np.sqrt(mse3), 2)))

fit4 = ExponentialSmoothing(train_data, seasonal_periods = seasonal_period, trend='add', seasonal='mul', damped=True, use_boxcox=True).fit()
fcast4 = fit4.forecast(len(predict_data)).rename('Multiplicative+damped')
mse4 = ((fcast3 - predict_data.actual_y) ** 2).mean()
print('The Root Mean Squared Error of additive damped trend, multiplicative seasonal of '+ 
      'period season_length={} and a Box-Cox transformation {}'.format(seasonal_period,round(np.sqrt(mse4), 2)))

fit3.fittedvalues.plot(style='--', marker = 'o', color='red')
fcast3.plot(style='--', marker='o', color='red', legend=True)
fit4.fittedvalues.plot(style='--',marker = 'o', color='green')
fcast4.plot(style='--', marker='o', color='green', legend=True)
```
